How to remove Missing/Null Values in a dataset. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../datasets/loan payments data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

```python
dataset.drop(columns=["paid_off_time"],inplace=True)
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

```python
dataset.dropna(inplace=True)
```
