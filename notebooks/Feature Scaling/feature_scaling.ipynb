```python
import pandas	as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =	pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv', header=None)
dataset.columns =	['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
print(dataset.head())
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x=dataset['MEDV'])
plt.show()
```

```python
sns.distplot(dataset['AGE'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(dataset[['AGE']])
```

```python
scaler.transform(dataset[['AGE']])
```

```python
dataset['scaled_AGE'] = pd.DataFrame(scaler.transform(dataset[['AGE']]), columns=['scaled_AGE'])
print(dataset.head(5))
```

```python
dataset.describe()
```

```python
sns.distplot(dataset['scaled_AGE'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))

# Before scaling
plt.subplot(1, 2, 1)
plt.title('Before Scaling')
sns.histplot(dataset['AGE'], kde=True)

# After scaling
plt.subplot(1, 2, 2)
plt.title('After Scaling')
sns.histplot(dataset['scaled_AGE'], kde=True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled = MinMaxScaler()
```

```python
scaled.fit(dataset[['AGE']])
```

```python
scaled_data =	scaled.transform(dataset[['AGE']])
print(scaled_data[:5])
```

```python
dataset['scaled_data'] = pd.DataFrame(scaled_data, columns=['scaled_AGE'])
dataset.head()
```

```python
plt.figure(figsize=(12, 5))

# Before scaling
plt.subplot(1, 2, 1)
plt.title('Before Scaling')
sns.histplot(dataset['AGE'], kde=True)

# After scaling
plt.subplot(1, 2, 2)
plt.title('After Scaling')
sns.histplot(dataset['scaled_data'], kde=True)

plt.tight_layout()
plt.show()
```
