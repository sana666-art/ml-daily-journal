```python
data = pd.DataFrame(dataset['Amount'])
data.head()
```

Remove Outlier in the Dataset

1. 	IQR Method

```python
dataset.shape
```

```python
q1 =	data['Amount'].quantile(0.25)
q3 =	data['Amount'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr #min range
upper_bound = q3 + 1.5 * iqr #max range

print("Lower Bound:",	lower_bound)
print("Upper Bound:",	upper_bound)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x=data['Amount'], data=dataset)
plt.axvline(lower_bound, color='red', linestyle='--')
plt.axvline(upper_bound, color='green', linestyle='--')
plt.show()
```

```python
New_dataset = dataset[dataset['Amount'] <= upper_bound]
```

```python
New_dataset.shape
```

```python
sns.boxplot(x=New_dataset['Amount'], data=New_dataset)
plt.show()
```

2. Z-score

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x='Amount', data=dataset)
```

```python
sns.distplot(dataset['Amount'])
plt.show()
```

```python
min_value = dataset['Amount'].mean() - (3*dataset['Amount'].std())

max_value = dataset['Amount'].mean() + (3*dataset['Amount'].std())

print("Min Value:", min_value)
print("Max Value:", max_value)
```

```python
# filter rows where Amount is between min_value and max_value (inclusive)
new_dataset = dataset[(dataset['Amount'] <= max_value) & (dataset['Amount'] >= min_value)]
```

```python
sns.boxplot(x=new_dataset['Amount'], data=new_dataset)
```

```python
z_score = (dataset['Amount'] - dataset['Amount'].mean()) / (dataset['Amount'].std())

z_score
```

```python
z_score > 3
```

```python
dataset[z_score] = z_score
```

```python
dataset['z_score'] = z_score

# filter rows with absolute z-score less than 3 (non-outliers)
dataset[dataset['z_score'].abs() < 3]
```
