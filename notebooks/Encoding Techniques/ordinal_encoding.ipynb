```python
import pandas as pd
```

```python
df = pd.DataFrame({
				'Size': ['Small', 'Medium', 'Large', 'Medium', 'Small',	'Large', 'Medium', 'Extra Small', 'Extra Large']})
df
```

```python
ordered_sizes = ['Extra Small', 'Small', 'Medium', 'Large', 'Extra Large']
```

```python
from	sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[ordered_sizes])
oe.fit(df[['Size']]) 
```

```python
oe.transform(df[['Size']])
```

```python
df['en_size'] = oe.transform(df[['Size']])
df
```

```python
oe = OrdinalEncoder(categories=[ordered_sizes])

df['Size_Encoded'] = oe.fit_transform(df[['Size']])

df
```

```python
dataset =	{'Low', 'Medium', 'High', 'Very High'}

ordered_dataset = {'Low': 0, 'Medium': 1, 'High': 2, 'Very High': 3}
```

```python
# Ensure 'dataset' column exists before mapping to avoid KeyError
if 'dataset' not in df.columns:
	# create a 'dataset' column by cycling through the ordered categories
	# so it has the same length as df
	categories = list(ordered_dataset.keys())
	df['dataset'] = (categories * ((len(df) // len(categories)) + 1))[:len(df)]

# Map to numeric codes using ordered_dataset
df['dataset_encoded'] = df['dataset'].map(ordered_dataset)

df[['dataset', 'dataset_encoded']]
```
