What is Polynomial Regression?

Polynomial Regression models a non-linear relationship between input and output by adding powers of features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2500000, 2300000, 2000000, 1700000, 1500000, 1400000])
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
# Predict price of a 3.5-year-old car
age = np.array([[3.5]])
age_poly = poly.transform(age)

predicted_price = model.predict(age_poly)
print("Predicted Price:", predicted_price[0])
```

```python
X_range = np.linspace(1, 6, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_pred = model.predict(X_range_poly)

plt.scatter(X, y, label="Actual Data")
plt.plot(X_range, y_pred, label="Polynomial Curve")
plt.xlabel("Car Age")
plt.ylabel("Price")
plt.legend()
plt.show()
```
